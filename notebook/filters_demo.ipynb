```python
# sobel filter Vertical
def simple_conv(imgFilter=None, picture=None):
    # Define a function that applies a convolution filter (imgFilter) on a grayscale image (picture)

    p_row, p_col = picture.shape  
    # Get the number of rows and columns in the image
    # picture.shape returns (height, width) for a 2D grayscale image

    k = imgFilter.shape[0]  
    # Get the size of the kernel/filter
    # Assumes a square kernel, so only one dimension is taken (e.g., 3x3 kernel → k=3)

    temp = list()  
    # Initialize an empty list to store the results of convolution for each position

    stride = 1  
    # Define the stride (step size) for moving the filter across the image
    # Here stride=1 means we move one pixel at a time

    final_cols = (p_col - k)//stride + 1  
    # Calculate the number of columns in the output image after convolution
    # Formula: (input_width - kernel_width) / stride + 1

    final_rows = (p_row - k)//stride + 1  
    # Calculate the number of rows in the output image after convolution
    # Formula: (input_height - kernel_height) / stride + 1

    for v_stride in range(final_rows):  
        # Loop over each possible vertical position of the filter
        # v_stride represents the top row index where the filter is applied

        for h_stride in range(final_cols):  
            # Loop over each possible horizontal position of the filter
            # h_stride represents the left column index where the filter is applied

            target_area_of_pic = picture[v_stride: v_stride + k, h_stride: h_stride + k]  
            # Extract a k x k patch from the image corresponding to the filter's current position
            # This is the region that will be multiplied element-wise with the filter

            z = sum(sum(imgFilter * target_area_of_pic))  
            # Element-wise multiplication of the filter with the patch
            # Then sum all values to get a single scalar output for this position
            # This is the basic convolution operation

            temp.append(z)  
            # Append the scalar result to the temp list

    resulant_image = np.array(temp).reshape(final_rows, final_cols)  
    # Convert the temp list to a numpy array
    # Reshape it into the size of the final output image (rows x cols)

    return resulant_image  
    # Return the convolved image
```
